### Desafio 07: Deixar os gráficos bonitos (pyplot)

```python
def percent(pct, allvals):
    absolute = int(round(pct/100.*np.sum(allvals)))
    return "{:.1f}%\n({:d})".format(pct, absolute)

fig, graph = plt.subplots(figsize=(10, 10), subplot_kw={"aspect": "equal"})

wedges, texts, autotexts = graph.pie(dados["tempo"].value_counts(),
    textprops={"color": "w"},
    autopct=lambda val: percent(val, dados["tempo"].value_counts())
)

graph.legend(wedges, dados["tempo"].unique(),
    title="Horas",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1),
    fontsize=12
)

plt.setp(autotexts, size=16, weight="bold")

graph.set_title("Tempo Decorrido", fontdict={"fontsize": 24})

plt.show()
```

### Desafio 02: Melhorar a vizualização

```python
dados.rename(columns={"droga": "composto"}, inplace=True)
```

```python
cod_compostos = dados["composto"].value_counts().index[0:5]

plt.figure(figsize=(10, 8))

ax = sns.countplot(
    x="composto",
    data=dados.query("composto in @cod_compostos"),
    order=cod_compostos
)

ax.set_title("Top 5 compostos mais testados", fontsize=20)
ax.set_ylabel("Quantidade de Testes", fontsize=16)
ax.set_xlabel("Composto", fontsize=16)
ax.tick_params(labelsize=14)

for patch in ax.patches:
    height = patch.get_height()
    ax.text(
        patch.get_x() + patch.get_width() / 2,
        height + 1,
        height,
        ha="center"
    )

plt.show()
```

### Desafio 03: Plotar os histogramas com Seaborn

```python
plt.figure(figsize=(10, 8))

ax = sns.histplot(
    data=dados.loc[:, "g-0":"g-771"].describe().T,
    x="mean",
    bins=30
)

ax.set_title("Distribuição de médias de g-X", fontsize=20)
ax.set_ylabel("Probabilidade (%)", fontsize=16)
ax.set_xlabel("Média", fontsize=16)
ax.tick_params(labelsize=14)

for patch in ax.patches:
    height = patch.get_height()
    ax.text(
        patch.get_x() + patch.get_width() / 2,
        height + 1,
        height,
        ha="center"
    )

plt.show()
```
