## Aula 02

### Alterando dados de uma coluna

```python
mapa = {
    "droga": "composto"
}

dados.rename(columns=mapa, inplace=True)

dados.head()
```

### Criando gráficos com Seaborn

```python
cod_compostos = dados["composto"].value_counts().index[0:5]
```

```python
import seaborn as sns

sns.set()
plt.figure(figsize=(10, 8))

sns.countplot(x="composto", data=dados.query("composto in @cod_compostos"), order=cod_compostos)
```
